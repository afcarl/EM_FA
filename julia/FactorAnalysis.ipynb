## Generate data from the Factor Analysis model (Gaussian)

$x_i \sim \mathcal{N}(0, I)$

$y_i|x_i \sim \mathcal{N}(Cx_i, R)$

where $R$ is diagonal. $C$ is called the *factor loadings* matrix, and $R$ is often referred to as the *uniqueness*.

```julia
q = 2; # Dimensionality of the latent space
p = 20; # Dimensionality of the observation space
N = 10000; # number of i.i.d. samples

CTrue = randn(p,q);
RTrue = diagm(0.01 + rand(p));
xTrue = randn(q, N);
v = randn(p, N);
y = CTrue * xTrue + RTrue * v;
```

Marginal distribution has a simple decomposition of covariance matrix.

$y_i \sim \mathcal{N}(0, CC^T + R)$

```julia
using PyPlot
Sigma_yy = y * y' / N;
ax1 = subplot(1,2,1); matshow(Sigma_yy, fignum=false); title("Empirical covariance")
ax2 = subplot(1,2,2); matshow(CTrue*CTrue' + RTrue, fignum=false); title("Theoretical cov")
```

# Let's do some Expectation-Maximization (EM)

```julia
# Initialize
qGuess = 2;
srand(20140805); # fix the random seed
C0 = randn(p,qGuess);
R0 = diagm(diag(Sigma_yy));
C = C0; R = R0;
```

```julia
function EM_iteration(R, C, Sigma_yy)
    # E-step
    # L = inv(eye(qGuess) + C'*(R\C))
    L = eye(qGuess) - C' * ((R + C * C') \ C);
    # Linv = eye(qGuess) + C'*(R\C);
    # Rinv = inv(R); # dont' panic, R is a diagonal matrix
    Rinv = diagm(1 ./ diag(R));
    delta = L * C' * Rinv
    # delta = Linv \ (C' * Rinv)
    
    # M-step
    C = Sigma_yy * delta' / (L + delta * Sigma_yy * delta');
    #R = diagm(diag(Sigma_yy - Sigma_yy * delta' * ((L + delta*Sigma_yy*delta')\delta) * Sigma_yy));
    R = diagm(diag(Sigma_yy - Sigma_yy * delta' * C'));
    return R, C
end
```

```julia
maxIteration = int(1e4);
residual = zeros(Float64, maxIteration);
@time begin
    for k = 1:maxIteration
        R, C = EM_iteration(R, C, Sigma_yy);
        residual[k] = norm(Sigma_yy - (C*C' + R))
        if residual[k] < 1e-2
            break;
        end
    end
end
println(residual[end])
```

```julia
semilogy(residual); xlabel("iteration"); ylabel("residual error"); title("convergence rate")
```

```julia
plot(diag(R), label="inferred"); plot(diag(RTrue), label="true");
legend(loc="upper right",fancybox="true"); title("independent noise")
```

# Let's do ML without EM

Maximize $\log p(y|C, R) = -\log|CC^T+R| - \mathrm{tr}[\Sigma_{yy}(CC^T+R)^{-1}]$

Derivatives:

$\frac{\partial \log p(y|C, R)}{\partial C} = -(CC^T+R)^{-1}C - (CC^T+R)^{-1} \Sigma_{yy} (CC^T+R)^{-1} C$

$\frac{\partial \log p(y|C, R)}{\partial C} = -\left((CC^T+R)^{-1} + (CC^T+R)^{-1} \Sigma_{yy} (CC^T+R)^{-1}\right) \circ \mathbf{I}$

```julia
using Optim
```

```julia
function cost(x::Vector)
    # unpack C and R
    C = reshape(x[1:qGuess*p], p, qGuess)
    R = diagm(exp(x[qGuess*p+1:end]));
    CCR = C*C'+R;
    logdet(CCR) + trace(Sigma_yy / CCR);
end
```

```julia
cost([C[:], log(diag(R))])
```

```julia
x = optimize(cost, [C0[:], log(diag(R0))], iterations=5000);
```

```julia
C2 = reshape(x.minimum[1:qGuess*p], p, qGuess);
R2 = diagm(exp(x.minimum[qGuess*p+1:end]));
println(cost([C2[:], log(diag(R2))]))
```

```julia
norm(Sigma_yy - (C2*C2' + R2))
```

```julia
plot(diag(R2), label="inferred"); plot(diag(RTrue), label="true");
legend(loc="upper right",fancybox="true"); title("independent noise")
```

## Closed form ML

$C_{new} := R^{1/2} U_t (\Lambda_t - \mathbf{I})^{1/2} V$

where

$R^{-1/2} \Sigma_{yy} R^{-1/2} = U \Lambda V$ is the SVD, and

$\Lambda_t$ and $U_t$ correspond to the singular values larger than 1.

$R_{new} := (\Sigma_{yy} - C_{new}C_{new}^T) \circ \mathbf{I}$

```julia
sqrtR = diagm(sqrt(diag(R)));
sqrtRinv = diagm(1 ./ (diag(sqrtR)));
U, L, V = svd(sqrtRinv * Sigma_yy * sqrtRinv);
Lt = L[1:qGuess]
Lt = Lt[Lt .> 1.0];
Ut = U[:, 1:length(Lt)];
Cnew = sqrtR * Ut * diagm(sqrt(Lt - 1))
Rnew = diagm(diag(Sigma_yy - Cnew * Cnew'))
```

```julia
function ECME2_iteration(R, C, Sigma_yy)
    sqrtR = diagm(sqrt(diag(R)));
    sqrtRinv = diagm(1 ./ (diag(sqrtR)));
    U, L, V = svd(sqrtRinv * Sigma_yy * sqrtRinv);
    Lt = L[1:qGuess]
    Lt = Lt[Lt .> 1.0];
    Ut = U[:, 1:length(Lt)];
    Cnew = sqrtR * Ut * diagm(sqrt(Lt - 1))
    Rnew = diagm(diag(Sigma_yy - Cnew * Cnew'))
    return Rnew, Cnew
end
```

```julia
maxIteration = int(100);
residual = zeros(Float64, maxIteration);
R = R0; C = C0;
@time begin
    for k = 1:maxIteration
        R, C = ECME2_iteration(R, C, Sigma_yy);
        residual[k] = norm(Sigma_yy - (C*C' + R))
        if residual[k] < 1e-2
            break;
        end
    end
end
println(residual[end])
```

```julia
plot(residual); xlabel("iteration"); ylabel("residual error"); title("convergence rate")
```
